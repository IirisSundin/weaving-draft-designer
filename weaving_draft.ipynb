```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
# Shafts and the size of the fabric in the visualization
varret = 16 # Number of shafts (varsien määrä)
l = 100 # width; number of warp threads for visualization (leveys: loimilankojen määrä)
h = 100 # height; number of weft threads for visualization (korkeus: kudelankojen määrä)

# create a random number generator for randomly created patterns
rng = np.random.default_rng(872)
```

```python
# Threading (niisintä)
# Point draw
niisinta = np.zeros((varret, l)) # niisinta = threading
for i in np.arange(l):
    niisinta[np.abs(i%((varret-1)*2)-(varret-1)), i] = 1
```

```python
plt.imshow(niisinta, cmap='Greys',  interpolation='nearest')
plt.title('Threading')
```

```python
# Treadling (poljenta)

poljenta = np.zeros((h, varret)) # poljenta = treadling
for i in np.arange(h):
    poljenta[i, varret-i%varret-1] = 1
poljenta = niisinta.transpose()
plt.imshow(poljenta, cmap='Greys',  interpolation='nearest')
plt.title('Treadling')
```

```python
# Tie-up (sidos)

# Initialize a random tie-up using a base vector that is shifted for different shafts 
# to achieve a typical twill-like pattern.
basevec = np.array([0,1,1,0,1,1,0,0,0,0,1,1,0, 1, 0, 1])

sidos = np.zeros((varret, varret))
for i in np.arange(varret):
    for shift in np.arange(varret):
        sidos[varret-(i+shift)%varret-1, i] = basevec[shift]

# For designing a random pattern, permutate the columns of the tie-up randomly
permutation = rng.permutation(np.arange(varret))
sidos = sidos[:,permutation]
```

```python
# Drawdown / the resulting pattern (kuvio)
kuvio = np.zeros((h,l))
for i in np.arange(h):
    for j in np.arange(l):
        kuvio[i,j] = sidos[poljenta[i,:]!=0, niisinta[:,j]!=0]
```

```python
plt.imshow(sidos, cmap='Greys',  interpolation='nearest')
plt.title('Tie-up')
```

```python
plt.imshow(kuvio, cmap='Greys',  interpolation='nearest')
plt.title('Drawdown: \n A visualization of the resulting pattern in a woven fabric')
```
